```python
import numpy as np

# 1-D array of 20 evenly spaced values between 0 and 1
linspace_arr = np.linspace(0, 1, 20)
# 5x5 identity matrix
identity_matrix = np.eye(5)
# 3x4 matrix of random integers between 1 and 100
rand_ints = np.random.randint(1, 101, (3, 4)).astype(np.int64)

print(f"linspace -> shape: {linspace_arr.shape}  dtype: {linspace_arr.dtype}")
print(f"identity -> shape: {identity_matrix.shape}  dtype: {identity_matrix.dtype}")
print(f"rand_ints -> shape: {rand_ints.shape}  dtype: {rand_ints.dtype}")
```

```python
arr = np.arange(1, 25)
print(f"Original  : {arr.shape}")

arr_3d = arr.reshape(2, 3, 4)
print(f"3-D       : {arr_3d.shape}")

arr_2d = arr_3d.reshape(6, 4)
print(f"2-D       : {arr_2d.shape}")

flattened = arr_2d.flatten()
print(f"Flattened : {flattened.shape}")
```

```python
matrix = np.random.rand(6, 6)
every_other = matrix[0:5:2, :]
diagonal = np.diag(matrix)
binary_matrix = np.where(matrix > 0.7, 1, 0)

print(f"Every-other rows -> shape: {every_other.shape}")
print(f"Diagonal         -> shape: {diagonal.shape}")
print(f"Binary matrix    -> shape: {binary_matrix.shape} values: {np.unique(binary_matrix)}")
```

```python
A = np.random.rand(4, 3)
B = np.random.rand(4, 3)

v_stacked = np.vstack((A, B))
h_stacked = np.hstack((A, B))
split_A, split_B = np.vsplit(v_stacked, 2)

print(f"vstack -> shape: {v_stacked.shape}")
print(f"hstack -> shape: {h_stacked.shape}")
print(f"split[0] equals A: {np.array_equal(split_A, A)}")
print(f"split[1] equals B: {np.array_equal(split_B, B)}")
```

```python
col_vec = np.arange(4).reshape(4, 1)
row_vec = np.arange(5).reshape(1, 5)
res1 = col_vec + row_vec

mat = np.random.rand(4, 3)
arr_3 = np.random.rand(3)
res2 = mat + arr_3

# Comment: These succeed because NumPy "stretches" dimensions of size 1 
# to match the larger array, aligning them from the right.
print(f"(4,1) + (1,5) -> shape: {res1.shape}")
print(f"(4,3) + (3,)  -> shape: {res2.shape}")
```

```python
X = np.random.randn(100, 5)
X_norm = (X - np.mean(X, axis=0)) / np.std(X, axis=0)

print(f"Column means: {np.mean(X_norm, axis=0).round(2)}")
print(f"Column stds : {np.std(X_norm, axis=0).round(2)}")
```

```python
X_min = np.min(X, axis=0)
X_max = np.max(X, axis=0)
X_scaled = (X - X_min) / (X_max - X_min)

# Comment: If a test point is outside the training range, 
# it will land outside the [0, 1] interval.
print(f"Min per column: {np.min(X_scaled, axis=0).round(2)}")
print(f"Max per column: {np.max(X_scaled, axis=0).round(2)}")
```

```python
W = np.random.randn(4, 3)
X = np.random.randn(3, 10)
b = np.random.randn(4, 1)

Z = W @ X
Z_plus_b = Z + b

print(f"W shape   : {W.shape}")
print(f"X shape   : {X.shape}")
print(f"Z shape   : {Z.shape}")
print(f"Z+b shape : {Z_plus_b.shape}")
```

```python
v1 = np.random.randn(128)
v2 = np.random.randn(128)
cos_sim = np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))

X = np.random.randn(10, 128)
X_norm = X / np.linalg.norm(X, axis=1, keepdims=True)
pairwise_sim = X_norm @ X_norm.T

print(f"Cosine similarity (two vectors): {cos_sim:.3f}")
print(f"Pairwise matrix shape          : {pairwise_sim.shape}")
print(f"Diagonal values                : {np.diag(pairwise_sim).round(1)}")
```

```python
data = np.random.randn(100, 4)
centered_data = data - np.mean(data, axis=0)
cov_matrix = np.cov(centered_data, rowvar=False)
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

idx = np.argsort(eigenvalues)[::-1]
top_2_vectors = eigenvectors[:, idx[:2]]
projected_data = centered_data @ top_2_vectors
var_explained = (eigenvalues[idx[:2]] / np.sum(eigenvalues)) * 100

print(f"Covariance matrix shape : {cov_matrix.shape}")
print(f"Eigenvalues (sorted)    : {eigenvalues[idx].round(2)}")
print(f"Projected data shape    : {projected_data.shape}")
print(f"Variance explained      : {var_explained.round(1)}%")
```
