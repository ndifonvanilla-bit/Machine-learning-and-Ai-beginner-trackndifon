```python
import pandas as pd

# Load the data
df_trans = pd.read_csv('transactions.csv')


# Convert 'amount' to numeric (handles $, commas, spaces, etc.)
df_trans['amount'] = pd.to_numeric(df_trans['amount'], errors='coerce')

# Optional: look at what we have now
print("Data types after cleaning:")
print(df_trans.dtypes)
print("\nMissing values in amount:", df_trans['amount'].isna().sum())

# ────────────────────────────────────────────────
# Task 1: Descriptive statistics
# ────────────────────────────────────────────────
print("\nDescriptive Statistics:")
print(df_trans.describe())

# ────────────────────────────────────────────────
# Task 2: Column with highest average (mean)
# ────────────────────────────────────────────────
numeric_desc = df_trans.describe()

if 'mean' in numeric_desc.index:
    highest_mean_col = numeric_desc.loc['mean'].idxmax()
    highest_mean_value = numeric_desc.loc['mean', highest_mean_col]
    print(f"\nColumn with highest average: {highest_mean_col}")
    print(f"   → Mean value: {highest_mean_value:.2f}")
else:
    print("\nNo numeric columns found — check if 'amount' converted correctly.")

# ────────────────────────────────────────────────
# Task 3: Column with highest standard deviation
# ────────────────────────────────────────────────
if 'std' in numeric_desc.index:
    highest_std_col = numeric_desc.loc['std'].idxmax()
    highest_std_value = numeric_desc.loc['std', highest_std_col]
    print(f"Column with highest variability (std): {highest_std_col}")
    print(f"   → Std value: {highest_std_value:.2f}")
else:
    print("No std row available — numeric conversion needed.")

# ────────────────────────────────────────────────
# Task 4: Business interpretation
# ────────────────────────────────────────────────
print("\nBusiness Insight:")
print("• The column with the highest average likely represents the most valuable/expensive transaction type or region.")
if 'amount' in df_trans.columns:
    print("• High average in 'amount' suggests premium customers, high-ticket items, or specific high-value categories/regions.")
print("• The column with highest standard deviation shows the most inconsistent/spiky behavior —")
print("  could indicate fraud risk, seasonal spikes, different customer segments, or unstable pricing.")
print("→ Investigate outliers in that column for deeper insights.")
```

```python
import pandas as pd

df = pd.read_csv('transactions.csv')

# Make sure 'amount' is numeric (very common source of follow-up errors)
df['amount'] = pd.to_numeric(df['amount'], errors='coerce')

# ────────────────────────────────────────────────
# Task 1: Transactions per customer
# ────────────────────────────────────────────────
print("Transactions per customer:")
print(df.groupby('customer_id').size())

# Alternative (more explicit column name):
# print(df.groupby('customer_id')['transaction_id'].size())   # if you have transaction_id
# or
# print(df['customer_id'].value_counts().sort_index())

# ────────────────────────────────────────────────
# Task 2: Avg & total per region
# ────────────────────────────────────────────────
region_stats = df.groupby('region')['amount'].agg(['mean', 'sum'])

print("\nRegion stats:")
print(region_stats)

# Nicer formatting (optional)
print("\nRegion stats (formatted):")
print(region_stats.rename(columns={'mean': 'Average Amount', 'sum': 'Total Revenue'}))

# ────────────────────────────────────────────────
# Task 3: Region with highest revenue
# ────────────────────────────────────────────────
if 'sum' in region_stats.columns:
    highest_revenue_region = region_stats['sum'].idxmax()
    highest_revenue_value  = region_stats['sum'].max()
    
    print(f"\nRegion with highest total revenue: {highest_revenue_region}")
    print(f"   → Total revenue: {highest_revenue_value:,.2f}")
else:
    print("\n'sum' column not found — check if 'amount' is numeric.")
```

```python
# 1. Convert usage_gb to numeric (just in case there are typos in the Excel)
df['usage_gb'] = pd.to_numeric(df['usage_gb'], errors='coerce')

# 2. Group by subscription_type using the correct column name 'usage_gb'
usage_stats = (
    df.groupby('subscription_type')['usage_gb']
    .agg(['mean', 'max', 'min'])
    .reset_index()
    .sort_values('mean', ascending=False)
)

print("\nNetwork usage by subscription type (sorted by average):")
print(usage_stats)
```

```python
customers = pd.read_csv('customers.csv')
transactions = pd.read_csv('transactions.csv')

# Task 1: Inner join
inner = pd.merge(customers, transactions, on='customer_id', how='inner')

# Task 2: Left join
left = pd.merge(customers, transactions, on='customer_id', how='left')

# Task 3: Right join
right = pd.merge(customers, transactions, on='customer_id', how='right')

# Task 4: Compare & explain
print(f"Inner join rows: {len(inner)}")
print(f"Left join rows : {len(left)}")
print(f"Right join rows: {len(right)}")
print("\nExplanation:")
print("- Inner: only customers who made transactions")
print("- Left : all customers (including those with no transactions)")
print("- Difference = customers without any transactions")


14
```

```python
customers = pd.read_csv('customers.csv')
transactions = pd.read_csv('transactions.csv')

# Task 1: Left join with indicator
merged = pd.merge(customers, transactions, on='customer_id', 
                  how='left', indicator=True)

# Task 2: Filter left_only
inactive = merged[merged['_merge'] == 'left_only']

# Task 3: Drop indicator column
inactive = inactive.drop(columns=['_merge'])

# Task 4: Count inactive customers
print(f"Inactive customers (no transactions): {len(inactive)}")
print("\nFirst 5 inactive customers:")
print(inactive.head())
```
