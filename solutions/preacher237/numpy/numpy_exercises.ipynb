```python
#Exercise 4
x = np.random.rand(4,3)
y = np.random.rand(4,3)
print("Matrix x:\n", x)
print("\nMatrix y:\n", y)

V = np.vstack((x,y))
print("\nVertical Stack shape:", V.shape)
H =np.hstack((x,y))
print("\nHorizonal Stack shape:", H.shape)
x_split, y_split =np.vsplit(V,2)

print("\nSplit x shape:", x_split.shape)
print("Split y shape:", y_split.shape)

assert np.allclose(x, x_split)
assert np.allclose(y, y_split)

print("\nAssertion passed: Split arrays match the originals.")

```
